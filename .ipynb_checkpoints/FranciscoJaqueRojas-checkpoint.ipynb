```python
data_country.index = data_country['Country Code']

data_country = data_country.drop('Country Code', axis=1)

data_country.head()
```

## ¿ha crecido la población mundial?
## ¿si ha habido crecimiento, en todos los países crece de la misma manera?

```python
fig, ax = plt.subplots(figsize=(18,8))
ax.plot(data_country.index, data_country['2021'] )
ax.plot(data_country.index, data_country['1960'], label='año 1960', color='orange' )

max_1960 = data_country['1960'].max()
max_2021 = data_country['2021'].max()

ax.axhline(max_1960, color='red', linestyle='--', alpha=0.5, linewidth=3, label='maxima de año 1960')
ax.axhline(max_2021, color='green', linestyle='--', alpha=0.5, linewidth=3, label='maxima de año 2021')           
       
    
ax.set_xlabel('Codigo pais')  
ax.set_ylabel('cantidad de poblacion')
ax.set_title('año 1960 y 2021')
ax.set_xlim(data_country.index[0], data_country.index[-1])
ax.legend()
```

## ¿cómo ha sido la demografía en estos últimos 10 años?

```python
fig, ax = plt.subplots(figsize=(18,8))
ax.plot(data_country.index, data_country['2021'] )
ax.plot(data_country.index, data_country['2011'], label='año 2011', color='orange' )

max_2011 = data_country['2011'].max()
max_2021 = data_country['2021'].max()

ax.axhline(max_2011, color='red', linestyle='--', alpha=0.5, linewidth=3, label='maxima de año 2011')
ax.axhline(max_2021, color='green', linestyle='--', alpha=0.5, linewidth=3, label='maxima de año 2021')           
       
    
ax.set_xlabel('Codigo pais')  
ax.set_ylabel('cantidad de poblacion')
ax.set_title('año 2011 y 2021')
ax.set_xlim(data_country.index[0], data_country.index[-1])
ax.legend()
```

#### Mi conclusión es la siguiente:  

Según los gráficos y análisis realizados, podemos decir que la población mundial ha crecido en los últimos años. Sin embargo, no necesariamente todos los países crecen de la misma manera, ya que algunos pueden tener tasas de crecimiento más altas que otros.

En cuanto a la demografía en los últimos 10 años, podemos observar la evolución de la población en 2011 y 2021 a través del segundo gráfico de línea, donde se compara la población de cada país en ambos años. A partir de esto, se podría analizar si la población ha aumentado o disminuido en cada país durante este período.
